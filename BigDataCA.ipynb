## Importing Data and Pre-Processing 

```python
#Import Libraries 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import data set from kaggle
boxing_data=pd.read_csv('C:/Big Data/bouts_out_new.csv')
```

```python
#Check the shape of the databoxing
boxing_data.shape
```

```python
#Checking for null values 
boxing_data.isnull().values.any()
```

```python
#Check the data frame 
boxing_data.head()
```

```python
#Pre-Processing Data 
#Remove Rows with Null Values & Removing Unwanted Columns 
boxing_data = boxing_data.drop(['judge1_A','judge1_B','judge2_A','judge2_B','judge3_A','judge3_B','decision'], axis=1)
boxing_data = boxing_data.dropna(how='any')
boxing_data.head()
```

## Change Datatypes


```python
#Some algorithims will only accept integers. To circumvent this issue, all data within the dataset will be converted to ints
boxing_data.head()
```

```python
#Stance for both fighters and result must be converted to integersstan
stance_map = {'win_A' : 1, 'win_B' : 2, 'draw' :0}
boxing_data['result'] = boxing_data['result'].map(stance_map)
boxing_data.head()
```

```python
stance_map = {'orthodox' : 0, 'southpaw' : 1,}
boxing_data['stance_A'] = boxing_data['stance_A'].map(stance_map)
boxing_data.head()
```

```python
stance_map = {'orthodox' : 0, 'southpaw' : 1,}
boxing_data['stance_B'] = boxing_data['stance_B'].map(stance_map)
boxing_data.head()
```

## Predictive Analysis - Splitting the Dataset 

```python
#Splitting the data
from sklearn.cross_validation import train_test_split #import training/test splitting method

training_names = ['age_A','age_B','height_A','height_B','reach_A','reach_B','weight_A','weight_B','won_A','won_B','lost_A','lost_B','drawn_A','drawn_B','kos_A','kos_B'] #columns for training 
predicted_names = ['result']  # predicted column for results
                  
X = boxing_data[training_names].values  # predictor columns
y = boxing_data[predicted_names].values # predictor 
split_test_size = 0.30

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=split_test_size, random_state=42)                        
```

```python
print("{0:0.2f}% in training set".format((len(X_train)/len(boxing_data.index)) * 100))
print("{0:0.2f}% in test set".format((len(X_test)/len(boxing_data.index)) * 100))
```

##  Naive Bayes 


```python
from sklearn.naive_bayes import GaussianNB
nb_model = GaussianNB()
nb_model.fit(X_train, y_train.ravel())
```

```python
#Accuracy of Training Data
nb_predict_train = nb_model.predict(X_train)
from sklearn import metrics

# training
print("Accuracy: {0:.2f}".format(metrics.accuracy_score(y_train, nb_predict_train)))
```

```python
#Accuracy of Testing Data
nb_predict_test = nb_model.predict(X_test)
from sklearn import metrics

#test
print("Accuracy: {0:.2f}".format(metrics.accuracy_score(y_test, nb_predict_test)))
```

```python
#Precision, F1 Score
print("Classification Report")
print(metrics.classification_report(y_test, nb_predict_test))
```

## Neural Network 


```python
#Import neural classifier 
from sklearn.neural_network import MLPClassifier

#scikt learn adopts the model construction and training interface 
mlpc = MLPClassifier()
mlpc.fit(X_train, y_train)

#Apply the model
boxing_data = mlpc.predict(X_test)

#Determining the accuracy 
score = metrics.accuracy_score(y_test, boxing_data) * 100
rounded_score = round(score, 1)
print("Neural Network Classifier Accuracy: {}%".format(rounded_score))
```
